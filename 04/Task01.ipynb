```python
import numpy as np
import matplotlib.pyplot as plt

mean_a = (1, 1)
cov_a = [[1, 0], [0, 1]]
n_a = 50
a = np.random.multivariate_normal(mean_a, cov_a, n_a)

mean_b = (2, 3)
cov_b = [[0.5, 0], [0, 0.5]]
n_b = 50
b = np.random.multivariate_normal(mean_b, cov_b, n_b)

mean_c = (4, 1)
cov_c = [[2, 0], [0, 2]]
n_c = 50
c = np.random.multivariate_normal(mean_c, cov_c, n_c)
```

```python
X = np.concatenate((a, b, c))
labels = np.ones(np.shape(X)[0])
labels[n_a : n_a + n_b] = 2
labels[n_a + n_b : n_a + n_b + n_c] = 3
indices = np.arange(len(labels))
np.random.shuffle(indices)
X_shuffled = X[indices]
labels_shuffled = labels[indices]
```

```python
figure = plt.figure(figsize=(6,6))
plt.scatter(X_shuffled[:,0], X_shuffled[:, 1], c = labels_shuffled)
```

```python
# Project into 5 and 50 dimensions
A_5 = np.random.rand(2, 5)
A_50 = np.random.rand(2, 50)
X_5 = np.matmul(X, A_5)
X_50 = np.matmul(X, A_50)
```

```python
from sklearn_lvq import GlvqModel, GrlvqModel, GmlvqModel

# Predict classification with models
glvq_5 = GlvqModel()
glvq_5.fit(X_5, labels_shuffled)
labels_predicted_glvq_5 = glvq_5.predict(X_5)

glvq_50 = GlvqModel()
glvq_50.fit(X_50, labels_shuffled)
labels_predicted_glvq_50 = glvq_50.predict(X_50)

grlvq_5 = GrlvqModel()
grlvq_5.fit(X_5, labels_shuffled)
labels_predicted_grlvq_5 = grlvq_5.predict(X_5)

grlvq_50 = GrlvqModel()
grlvq_50.fit(X_50, labels_shuffled)
labels_predicted_grlvq_50 = grlvq_50.predict(X_50)

gmlvq_5 = GmlvqModel()
gmlvq_5.fit(X_5, labels_shuffled)
labels_predicted_gmlvq_5 = gmlvq_5.predict(X_5)

gmlvq_50 = GmlvqModel()
gmlvq_50.fit(X_50, labels_shuffled)
labels_predicted_gmlvq_50 = gmlvq_50.predict(X_50)
```

```python
from sklearn_lvq.utils import plot2d

# Plot the transformation of the data
figure = plt.figure(figsize=(12,12))

plot2d(glvq_5, X_5, labels_shuffled, figure=0, title="GLVQ 5")
plot2d(glvq_50, X_50, labels_shuffled, figure=1, title="GLVQ 50")
plot2d(grlvq_5, X_5, labels_shuffled, figure=2, title="GRLVQ 5")
plot2d(grlvq_50, X_50, labels_shuffled, figure=3, title="GRLVQ 50")
plot2d(gmlvq_5, X_5, labels_shuffled, figure=4, title="GMLVQ 5")
plot2d(gmlvq_50, X_50, labels_shuffled, figure=5, title="GMLVQ 50")
```

```python
# Plot results
figure = plt.figure(figsize=(12,6))
figure.suptitle("GLVQ")
ax1 = figure.add_subplot(1,2,1)
ax1.scatter(X_shuffled[:,0], X_shuffled[:,1], c = labels_predicted_glvq_5)
ax1.title.set_text('5 dimensions')
ax2 = figure.add_subplot(1,2,2)
ax2.scatter(X_shuffled[:,0], X_shuffled[:,1], c = labels_predicted_glvq_50)
ax2.title.set_text('50 dimensions')

figure = plt.figure(figsize=(12,6))
figure.suptitle('GRLVQ')
ax1 = figure.add_subplot(1,2,1)
ax1.scatter(X_shuffled[:,0], X_shuffled[:,1], c = labels_predicted_grlvq_5)
ax1.title.set_text('5 dimensions')
ax2 = figure.add_subplot(1,2,2)
ax2.scatter(X_shuffled[:,0], X_shuffled[:,1], c = labels_predicted_grlvq_50)
ax2.title.set_text('50 dimensions')

figure = plt.figure(figsize=(12,6))
figure.suptitle('GMLVQ')
ax1 = figure.add_subplot(1,2,1)
ax1.scatter(X_shuffled[:,0], X_shuffled[:,1], c = labels_predicted_gmlvq_5)
ax1.title.set_text('5 dimensions')
ax2 = figure.add_subplot(1,2,2)
ax2.scatter(X_shuffled[:,0], X_shuffled[:,1], c = labels_predicted_gmlvq_50)
ax2.title.set_text('50 dimensions')
```

```python
# Plot accuracy of the models
figure = plt.figure(figsize=(12,6))
figure.suptitle('Accuracy')
ax1 = figure.add_subplot(1,2,1)
ax1.bar(['GLVQ', 'GRLVQ', 'GMLVQ'], [glvq_5.score(X_5, labels_shuffled), grlvq_5.score(X_5, labels_shuffled), gmlvq_5.score(X_5, labels_shuffled)])
ax1.title.set_text('5 dimensions')
ax2 = figure.add_subplot(1,2,2)
ax2.bar(['GLVQ', 'GRLVQ', 'GMLVQ'], [glvq_50.score(X_50, labels_shuffled), grlvq_50.score(X_50, labels_shuffled), gmlvq_50.score(X_50, labels_shuffled)])
ax2.title.set_text('50 dimensions')
```
