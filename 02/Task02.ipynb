```python
from sklearn.datasets import load_breast_cancer, load_diabetes, load_wine
from sklearn.decomposition import PCA
from sklearn.manifold import MDS, Isomap

# Load the data
breast_cancer = load_breast_cancer()
diabetes = load_diabetes()
wine = load_wine()

# Reduce dimensions with pca, mds and isomap
breast_cancer_pca = PCA(n_components=2).fit_transform(breast_cancer.data)
breast_cancer_mds = MDS(n_components=2).fit_transform(breast_cancer.data)
breast_cancer_isomap = Isomap(n_components=2).fit_transform(breast_cancer.data)

diabetes_pca = PCA(n_components=2).fit_transform(diabetes.data)
diabetes_mds = MDS(n_components=2).fit_transform(diabetes.data)
diabetes_isomap = Isomap(n_components=2).fit_transform(diabetes.data)

wine_pca = PCA(n_components=2).fit_transform(wine.data)
wine_mds = MDS(n_components=2).fit_transform(wine.data)
wine_isomap = Isomap(n_components=2).fit_transform(wine.data)
```

```python
import matplotlib.pyplot as plt

# Plot pca, lda, isomap
figure = plt.figure(figsize=(15, 15))
figure.suptitle('PCA, MDS, Isomap')

ax1 = figure.add_subplot(3, 3, 1)
ax1.scatter(breast_cancer_pca[:, 0], breast_cancer_pca[:, 1], c=breast_cancer.target)
ax1.set_title('PCA Breast Cancer')

ax2 = figure.add_subplot(3, 3, 2)
ax2.scatter(breast_cancer_mds[:, 0], breast_cancer_mds[:, 1], c=breast_cancer.target)
ax2.set_title('MDS Breast Cancer')

ax3 = figure.add_subplot(3, 3, 3)
ax3.scatter(breast_cancer_isomap[:, 0], breast_cancer_isomap[:, 1], c=breast_cancer.target)
ax3.set_title('Isomap Breast Cancer')

ax4 = figure.add_subplot(3, 3, 4)
ax4.scatter(diabetes_pca[:, 0], diabetes_pca[:, 1], c=diabetes.target)
ax4.set_title('PCA Diabetes')

ax5 = figure.add_subplot(3, 3, 5)
ax5.scatter(diabetes_mds[:, 0], diabetes_mds[:, 1], c=diabetes.target)
ax5.set_title('MDS Diabetes')

ax6 = figure.add_subplot(3, 3, 6)
ax6.scatter(diabetes_isomap[:, 0], diabetes_isomap[:, 1], c=diabetes.target)
ax6.set_title('Isomap Diabetes')

ax7 = figure.add_subplot(3, 3, 7)
ax7.scatter(wine_pca[:, 0], wine_pca[:, 1], c=wine.target)
ax7.set_title('PCA Wine')

ax8 = figure.add_subplot(3, 3, 8)
ax8.scatter(wine_mds[:, 0], wine_mds[:, 1], c=wine.target)
ax8.set_title('MDS Wine')

ax9 = figure.add_subplot(3, 3, 9)
ax9.scatter(wine_isomap[:, 0], wine_isomap[:, 1], c=wine.target)
ax9.set_title('Isomap Wine')

plt.show()
```
